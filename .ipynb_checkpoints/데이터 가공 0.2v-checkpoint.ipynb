- Kg당 가격(kgperprice)
  - 가격(sbidpric) / 거래단량(delngprut)
  - 거래단위(stdunitnewcode)가g(11)일경우 *1000, kg(12)일경우 그대로, ton(13)일경우 /1000
- 총거래가격(totalprice)
  - 거래량(delngQy) * 가격(sbidpric)
- 총량(totalweight)
  - 거래단량(delngprut) * 거래량(delngQy)
- 구품목코드(stdprdlstcode)
  - 현재 배추(1001)만
  - 추후 양파, 무, 마늘 등등 주요 민감 5개품목 관련해서도 진행
- 등급코드(stdqlitynewcode)
  - 일단 상(12)만 하는것으로 회의했었음
  - 하지만 작황이 좋으면 특(11) 안좋으면 보통(13) 또는 그 이외도 나올수 있기때문에 고려해봐야됨
- 구시장코드(whsalmrktcode)
  - 가락시장(110001)만

```python
df2018 = pd.read_csv("C:/crop_data/2018010120181231.csv")
df2017 = pd.read_csv("C:/crop_data/2017010120171231.csv")
df2016 = pd.read_csv("C:/crop_data/2016010120161231.csv")
df2015 = pd.read_csv("C:/crop_data/2015010120151231.csv")
```

```python
def customFunc(df):
    df = df[((df['stdunitnewcode']==11) | (df['stdunitnewcode']==12) | (df['stdunitnewcode']==13))
              &(df['stdprdlstcode']==1001) 
              &(df['whsalmrktcode']==110001)]
    
    #Kg당 가격(kgperprice)
    df['kgperprice'] = ''
    #총거래가격(totalprice)
    df['totalprice'] = ''
    #총량(totalweight)
    df['totalweight'] = ''
    
    df11 = df[(df['stdunitnewcode']==11)]
    df12 = df[(df['stdunitnewcode']==12)]
    df13 = df[(df['stdunitnewcode']==13)]
    for df11 in [df11]:
        df11['kgperprice'] = df11['sbidpric'] / df11['delngprut'] * 1000
    for df12 in [df12]:
        df12['kgperprice'] = df12['sbidpric'] / df12['delngprut']
    for df13 in [df13]:
        df13['kgperprice'] = df13['sbidpric'] / df13['delngprut'] / 1000
    df = pd.concat([df11, df12, df13])

    for df in [df]:
        #총거래가격(totalprice) = 거래량(delngQy) * 가격(sbidpric)
        df['totalprice'] = df['sbidpric'] * df['delngqy']
        #총량(totalweight) = 거래단량(delngprut) * 거래량(delngQy)
        df['totalweight'] = df['delngprut'] * df['delngqy']
    return df
```

```python
data = pd.concat([df['kgperprice'], df['stdspciesnewnm']], axis=1)
f, ax = plt.subplots(figsize=(14, 6))
fig = sns.boxplot(x='stdspciesnewnm', y='kgperprice', data=data)
```

```python
df = customFunc(df2018)
# day_cost2 = pd.DataFrame(df.loc[((df['stdspciesnewnm']=='월동배추')|(df['stdspciesnewnm']=='봄배추')|(df['stdspciesnewnm']=='김장(가을)배추')|(df['stdspciesnewnm']=='절임배추')|(df['stdspciesnewnm']=='저장배추')|(df['stdspciesnewnm']=='여름배추')|(df['stdspciesnewnm']=='고냉지배추'))
# day_cost2 = pd.DataFrame(df.loc[(df['stdspciesnewnm']=='김장(가을)배추')
day_cost2 = pd.DataFrame(df.loc[((df['stdqlitynewcode']=='11') | (df['stdqlitynewcode']=='12'))].groupby('delngde')['kgperprice'].mean())
plt.figure(num=None, figsize=(20, 6))
sns.pointplot(day_cost2.index,day_cost2['kgperprice'])
plt.show()
```
