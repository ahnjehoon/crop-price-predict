```python
url = 'http://apis.data.go.kr/B552895/openapi/service/OrgPriceAuctionService/getExactProdPriceList'
params = {
     'ServiceKey': ServiceKey  
    ,'pageNo': pageNo
    ,'numOfRows': numOfRows
    ,'prdlstCd': prdlstCd
}
```

```python
date_list = [d.strftime('%Y%m%d') for d in pd.date_range(year+'0101',year+'1231')]
error_list = []
zero_list = []
for v in date_list:
    params['delngDe'] = v
    params_encode = "&".join("%s=%s" % (k,v) for k,v in params.items())
    
    while(True):
        data = connectionXML(url, params_encode)
        if(data.find('totalcount')):
            break
        # data.find('returnauthmsg').text == 'HTTP_ERROR'
    totalcount = data.find('totalcount').text
    result_list = data.find_all('item')
    
    
    if(len(result_list) == 0):
#         zero_list.append(v)
        print(v, '날짜가 0임')
    elif(str(len(result_list)) == totalcount):
        getDataXML(data)
    else:
#         error_list.append(v)
        print(v,'날짜와 데이터가 맞지 않음')
```
