```python
# 필요한 변수들만 줄인다.
df2 = df1.loc[:,['경매일자', '거래단량', '거래량', '거래가격', '구산지코드', '구품종코드', '구품목코드', '등급코드', '단위명', '구시장코드']]
df2.head()
```

```python
n_data = df2.shape[0]
n_data
```

포장상태는 상관없고

가격/ 거래단량/ 거래량

#### 가격/거래단량 = kg당 가격


#### 거래단량 * 거래량 = 총 거래량 (단위 생각) -> kg 기준

```python
sns.countplot(df2['단위명'])
plt.show()
```

```python
df2['kg가격'] = df2['거래가격']/df2['거래단량']
df2['총거래량'] = df2['거래단량']*df2['거래량']
df2.head()
```

```python
# 가격/거래단량 = kg당 가격
# 거래단량 * 거래량 = 총 거래량 (단위 생각) -> kg 기준?
# 단위명 먼저 나눠서
for i in range(0,n_data-1) :
    if df2['단위명'][i] == 'g':
        df2['kg가격'][i] = df2['kg가격'][i]*1000
        df2['총거래량'][i] = df2['총거래량'][i]/1000
    elif df2['단위명'][i] == 'ton':
        df2['kg가격'][i] = df2['kg가격'][i]/1000
        df2['총거래량'][i] = df2['총거래량'][i]*1000
df2.head()
```
