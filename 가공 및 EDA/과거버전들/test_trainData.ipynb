포장상태는 상관없고

가격/ 거래단량/ 거래량

#### 가격/거래단량 = kg당 가격


#### 거래단량 * 거래량 = 총 거래량 (단위 생각) -> kg 기준

```python
n_data = df4.shape[0]
n_data
```

```python
sns.countplot(df4['단위명'])
plt.show()
```

```python
df4['kg가격'] = df4['거래가격']/df4['거래단량']
df4['총거래량'] = df4['거래단량']*df4['거래량']
df4.head()
```

```python
df4 = df4.reset_index()
df4.head(20)
```

```python
# 가격/거래단량 = kg당 가격
# 거래단량 * 거래량 = 총 거래량 (단위 생각) -> kg 기준?
# 단위명 먼저 나눠서
for i in range(0,n_data-1) :
    if df4['단위명'][i] == 'g':
        df4['kg가격'][i] = df4['kg가격'][i]*1000
        df4['총거래량'][i] = df4['총거래량'][i]/1000
    elif df4['단위명'][i] == 'ton':
        df4['kg가격'][i] = df4['kg가격'][i]/1000
        df4['총거래량'][i] = df4['총거래량'][i]*1000
df4.head()
```

```python
df4 = df4.drop(columns='index')
```

```python
df4.loc[df4['단위명']=='kg'].max()
```

```python
df4.to_csv('mid_2018_data.csv', header='infer', index=False, encoding='utf-8')
```

## 경매일자를 년, 월, 일로 나누기 

```python
df_2008 = pd.read_csv('mid_2008_data.csv', header='infer', encoding = 'utf-8')
df_2009 = pd.read_csv('mid_2009_data.csv', header='infer', encoding = 'utf-8')
df_2010 = pd.read_csv('mid_2010_data.csv', header='infer', encoding = 'utf-8')
df_2011 = pd.read_csv('mid_2011_data.csv', header='infer', encoding = 'utf-8')
df_2012 = pd.read_csv('mid_2012_data.csv', header='infer', encoding = 'utf-8')
df_2013 = pd.read_csv('mid_2013_data.csv', header='infer', encoding = 'utf-8')
df_2014 = pd.read_csv('mid_2014_data.csv', header='infer', encoding = 'utf-8')
df_2015 = pd.read_csv('mid_2015_data.csv', header='infer', encoding = 'utf-8')
df_2016 = pd.read_csv('mid_2016_data.csv', header='infer', encoding = 'utf-8')
df_2017 = pd.read_csv('mid_2017_data.csv', header='infer', encoding = 'utf-8')
df_2018 = pd.read_csv('mid_2018_data.csv', header='infer', encoding = 'utf-8')
```

```python
df = pd.read_csv('mid_2008_data.csv', header='infer', encoding = 'utf-8')
```

```python
df['경매일자'] = df['경매일자'].apply(lambda x : str(x))
df['연도'] = df['경매일자'].apply(lambda x : x[:4])
df['월'] = df['경매일자'].apply(lambda x : x[4:6])
df['일'] = df['경매일자'].apply(lambda x : x[6:])
df['연도'] = df['연도'].apply(lambda x : int(x))
df['월'] = df['월'].apply(lambda x : int(x))
df['일'] = df['일'].apply(lambda x : int(x))
df1 = df.groupby(['경매구분코드','월']).mean()
month_mean_1 = df1.loc[1][['kg가격','총거래량']]
month_mean_2 = df1.loc[2][['kg가격','총거래량']]

print(month_mean_1.head())
print(month_mean_2.head())
```

```python
def month_mean(df):
    df['경매일자'] = df['경매일자'].apply(lambda x : str(x))
    df['연도'] = df['경매일자'].apply(lambda x : x[:4])
    df['월'] = df['경매일자'].apply(lambda x : x[4:6])
    df['일'] = df['경매일자'].apply(lambda x : x[6:])
    df['연도'] = df['연도'].apply(lambda x : int(x))
    df['월'] = df['월'].apply(lambda x : int(x))
    df['일'] = df['일'].apply(lambda x : int(x))
    df1 = df.groupby(['경매구분코드','월']).mean()
    month_mean_1 = df1.loc[1][['kg가격','총거래량']]
    month_mean_2 = df1.loc[2][['kg가격','총거래량']] 
    return month_mean_1, month_mean_2
```

```python
month_2008_1, month_2008_2 = month_mean(df_2008)
month_2009_1, month_2009_2 = month_mean(df_2009)
month_2010_1, month_2010_2 = month_mean(df_2010)
month_2011_1, month_2011_2 = month_mean(df_2011)
month_2012_1, month_2012_2 = month_mean(df_2012)
month_2013_1, month_2013_2 = month_mean(df_2013)
month_2014_1, month_2014_2 = month_mean(df_2014)
month_2015_1, month_2015_2 = month_mean(df_2015)
month_2016_1, month_2016_2 = month_mean(df_2016)
month_2017_1, month_2017_2 = month_mean(df_2017)
month_2018_1, month_2018_2 = month_mean(df_2018)
```

```python
month_mean_1 = pd.concat([month_2008_1,month_2009_1,month_2010_1,month_2011_1,month_2012_1,month_2013_1,month_2014_1,month_2015_1,month_2016_1,month_2017_1,month_2018_1])
month_mean_2 = pd.concat([month_2008_2,month_2009_2,month_2010_2,month_2011_2,month_2012_2,month_2013_2,month_2014_2,month_2015_2,month_2016_2,month_2017_2,month_2018_2])
```

```python
month_mean_1.columns = ['kgPrice', 'totalAmount']
month_mean_2.columns = ['kgPrice', 'totalAmount']
#month_mean_all
```

```python
month_mean_1.to_csv('mean_of_month_1.csv', header='infer', index=False)
month_mean_2.to_csv('mean_of_month_2.csv', header='infer', index=False)
```
